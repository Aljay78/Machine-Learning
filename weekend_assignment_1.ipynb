Problem 10

```python
# Problem 10
original_price = 2500                            # --- Original Price per gallon of palm oil
discount_rate = 0.10                             # --- 10% discount rate

discount_amount = original_price * discount_rate # --- Calculating discount
final_price = original_price - discount_amount   # --- Substrating discount from original price

print("Discount Amount:", discount_amount, "LRD")
print("Final price after discount:", final_price, "LRD")
```
