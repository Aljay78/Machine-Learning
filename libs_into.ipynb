Numpy Installation

```python
import numpy as np
print(np.__version__)
print("\n")

# Daily gari prices at Gobachop Market (LRD per bag)
prices = np.array([850, 900, 875, 920, 860, 910, 895])

print('Prices:',         prices)
print('Mean Price:',     np.mean(prices).round(2))
print('Median Price:',   np.median(prices))
print('Highest Price:',  np.max(prices))
print('Lowest Price:',   np.min(prices))
print('Std Dev:',        np.std(prices).round(2))
```

Pandas: Code Block 2.1

```python
import pandas as pd
url = "https://raw.githubusercontent.com/Aljay78/Machine-Learning/refs/heads/main/gari_price_data.csv"
df = pd.read_csv(url)
print("\n")
print(df.head())                         # first 5 row
print(df.iloc[10:26,:])                  # row 10 t0 25
print('\nShape:', df.shape)              # rows x columns
print('\nColumns:', df.columns.tolist())
print('\nDescribe:\n', df.describe().round(2))

print("\n")
print(df.tail())                         # last 5 row

#print(df)                                # entire row
```

Matplotlib: Code Block 3.1

```python
import matplotlib.pyplot as plt

days = list(range(1, 8))
prices = [850, 900, 875, 920, 860, 910, 895]

plt.figure(figsize=(8, 4))
plt.plot(days, prices, marker='o', color='darkorange', linewidth=2)
plt.title('Gari Price Trend - Gobachop Market (LRD)')
plt.xlabel('Day')   ;  plt.ylabel('Price Per Bag (LRD)')
plt.grid(True)      ;  plt.tight_layout()   ;    plt.show()
```
